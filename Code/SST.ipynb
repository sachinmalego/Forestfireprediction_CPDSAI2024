```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
sst = pd.read_table('../datasets/sst_weekly.txt', 
                    header=None, 
                    sep='\s+', 
                    index_col=0, 
                    parse_dates=True, 
                    skiprows=4, 
                    skip_blank_lines=True,
                    names=['date', 'nino12_sst', 'nino12_ssta', 'nino3_sst', 'nino3_ssta', 'nino34_sst', 'nino34_ssta', 'nino4_sst', 'nino4_ssta'])

sst.head()
```

```python
# Resample weekly data to daily data, and forward fill the missing values
sst_daily = sst.resample('D').ffill()

sst_daily.head()
```

```python
# Filter sst_daily by year 2007
sst_daily_2007 = sst_daily['2007']

# Filter sst_daily by year 2012
sst_daily_2012 = sst_daily['2012']

# Write to csv
sst_daily_2007.to_csv('../datasets/sst_daily_2007.csv')
sst_daily_2012.to_csv('../datasets/sst_daily_2012.csv')
```
